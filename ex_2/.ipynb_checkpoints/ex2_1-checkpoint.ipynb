```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Const to the part
SIZE = 100000
loc1 = 2
loc2 = 5
scale1=2
scale2=5

dict_with_data = {
    "r_nor":np.random.normal(size=SIZE),
    "r_nor_loc_{}_scale_{}".format(loc1, scale1):np.random.normal(size=SIZE, scale=scale1, loc=loc1),
    "r_nor_loc_{}_scale_{}".format(loc2, scale2):np.random.normal(size=SIZE, scale=scale2, loc=loc2)
}
```

```python
# convert to the data frame object
df_nor=pd.DataFrame.from_dict(dict_with_data)

# it allow to use all the functions that 
# are deliver in the pandas package
```

```python
# melt data 
```

```python
df_nor_melt = df_nor.melt()
```

```python
# show the hist
bins1 = 128

g_nor_melt=sns.FacetGrid(df_nor_melt, col="variable", sharex=True, col_wrap=3)
g_nor_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_nor_melt.fig.suptitle("bins={}, cumulative False".format(bins1), y =1.05)
# show the diffrence with the cumulative mode, rest the same
g_nor_melt=sns.FacetGrid(df_nor_melt, col="variable", sharex=True, col_wrap=3)
g_nor_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_nor_melt.fig.suptitle("bins={}, cumulative True".format(bins1), y =1.05)
```

First 3 plots are histogram plots for different values of scale factor. From those plots we can get information about the distribution of the (in this case numerical, it could be categorical as well) data. 
From the next 3 plots we can get information about the cumulative cases over the range of cases. So basically it tells what is a percentage of the total number of cases.

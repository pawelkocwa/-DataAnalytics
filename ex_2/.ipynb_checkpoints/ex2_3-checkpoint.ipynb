```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import logit, expit
```

```python
# Poisson distribution
```

```python
# Const to the part This is a code from ex2_1
SIZE = 100000
int1 = 0
int2 = 1
int3 = 2
int4 = 5

dict_with_data = {
    "p_int_{}".format(int1):np.random.poisson(lam=int1, size=SIZE),
    "p_int_{}".format(int2):np.random.poisson(lam=int2, size=SIZE),
    "p_int_{}".format(int3):np.random.poisson(lam=int3, size=SIZE),
    "p_int_{}".format(int4):np.random.poisson(lam=int4, size=SIZE)
}
# convert to the data frame object
df_pois=pd.DataFrame.from_dict(dict_with_data)
# melt data
df_pois_melt = df_pois.melt()
```

```python
# show the hist for the explit data
bins1 = 128

g_pois_melt=sns.FacetGrid(df_pois_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_pois_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_pois_melt.fig.suptitle("Poisson\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_pois_melt=sns.FacetGrid(df_pois_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_pois_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_pois_melt.fig.suptitle("Poisson\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# show the hist for the explit data
bins1 = 64

g_pois_melt=sns.FacetGrid(df_pois_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_pois_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_pois_melt.fig.suptitle("Poisson\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_pois_melt=sns.FacetGrid(df_pois_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_pois_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_pois_melt.fig.suptitle("Poisson\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# show the hist for the explit data
bins1 = 256

g_pois_melt=sns.FacetGrid(df_pois_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_pois_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_pois_melt.fig.suptitle("Poisson\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_pois_melt=sns.FacetGrid(df_pois_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_pois_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_pois_melt.fig.suptitle("Poisson\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# The expit function, also known as the logistic sigmoid function, 
# is defined as expit(x) = 1/(1+exp(-x)). It is the inverse of the logit function.

df_expit = expit(df_pois).melt()
```

```python
# show the hist for the explit data
bins1 = 128

g_pois_melt=sns.FacetGrid(df_expit, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_pois_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_pois_melt.fig.suptitle("Poisson expit\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_pois_melt=sns.FacetGrid(df_expit, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_pois_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_pois_melt.fig.suptitle("Poisson expit\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# show the hist for the explit data
bins1 = 64

g_pois_melt=sns.FacetGrid(df_expit, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_pois_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_pois_melt.fig.suptitle("Poisson expit\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_pois_melt=sns.FacetGrid(df_expit, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_pois_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_pois_melt.fig.suptitle("Poisson expit\nbins={}, cumulative True".format(bins1), y =1.15)
```

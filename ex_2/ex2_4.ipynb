```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import logit, expit
```

```python
# Beta distribution
```

```python
# Const to the part This is a code from ex2_1
SIZE = 100000
beta1 = 1
beta2 = 2
beta3 = 3
beta4 = 4
alpha1 = 4
alpha2 = 3
alpha3 = 2
alpha4 = 1

dict_with_data = {
    "a_{}_b_{}_dist".format(alpha1, beta1):np.random.beta(a=alpha1, b=beta1, size=SIZE),
    "a_{}_b_{}_dist".format(alpha2, beta2):np.random.beta(a=alpha2, b=beta2, size=SIZE),
    "a_{}_b_{}_dist".format(alpha3, beta3):np.random.beta(a=alpha3, b=beta3, size=SIZE),
    "a_{}_b_{}_dist".format(alpha4, beta4):np.random.beta(a=alpha4, b=beta4, size=SIZE)
}
# convert to the data frame object
df_beta=pd.DataFrame.from_dict(dict_with_data)
# melt data
df_beta_melt = df_beta.melt()
```

```python
# show the hist for the explit data
bins1 = 128

g_beta_melt=sns.FacetGrid(df_beta_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_beta_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_beta_melt.fig.suptitle("Beta\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_beta_melt=sns.FacetGrid(df_beta_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_beta_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_beta_melt.fig.suptitle("Beta\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# show the hist for the explit data
bins1 = 64

g_beta_melt=sns.FacetGrid(df_beta_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_beta_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_beta_melt.fig.suptitle("Beta\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_beta_melt=sns.FacetGrid(df_beta_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_beta_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_beta_melt.fig.suptitle("Beta\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# The expit function, also known as the logistic sigmoid function, 
# is defined as expit(x) = 1/(1+exp(-x)). It is the inverse of the logit function.

df_expilt = expit(df_beta).melt()
```

```python
# show the hist for the explit data
bins1 = 128

g_beta_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_beta_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_beta_melt.fig.suptitle("Beta expit\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_beta_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_beta_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_beta_melt.fig.suptitle("Beta expit\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# show the hist for the explit data
bins1 = 64

g_beta_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_beta_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_beta_melt.fig.suptitle("Beta expit\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_beta_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_beta_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_beta_melt.fig.suptitle("Beta expit\nbins={}, cumulative True".format(bins1), y =1.15)
```

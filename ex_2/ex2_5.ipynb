```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import logit, expit
```

```python
# Log-normal distribution
```

```python
# Const to the part This is a code from ex2_1
SIZE = 100000
mean1 = 0
mean2 = 1
mean3 = 3
mean4 = 5
sigma1 = 1
sigma2 = 0.9
sigma3 = 0.5
sigma4 = 0.3

dict_with_data = {
    "mean_{}_sigma_{}".format(mean1, sigma1):np.random.lognormal(mean=mean1, sigma=sigma1, size=SIZE),
    "mean_{}_sigma_{}".format(mean2, sigma2):np.random.lognormal(mean=mean2, sigma=sigma2, size=SIZE),
    "mean_{}_sigma_{}".format(mean3, sigma3):np.random.lognormal(mean=mean3, sigma=sigma3, size=SIZE),
    "mean_{}_sigma_{}".format(mean4, sigma4):np.random.lognormal(mean=mean4, sigma=sigma4, size=SIZE)
}
# convert to the data frame object
df_lognor=pd.DataFrame.from_dict(dict_with_data)
# melt data
df_lognor_melt = df_lognor.melt()
```

```python
# show the hist for the explit data
bins1 = 128

g_lognor_melt=sns.FacetGrid(df_lognor_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_lognor_melt.map(plt.hist,"value",cumulative=False, bins=bins1)
g_lognor_melt.fig.suptitle("Log-normal\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_lognor_melt=sns.FacetGrid(df_lognor_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_lognor_melt.map(plt.hist,"value",cumulative=True, bins=bins1)
g_lognor_melt.fig.suptitle("Log-normal\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# show the hist for the explit data
bins1 = 64

g_lognor_melt=sns.FacetGrid(df_lognor_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_lognor_melt.map(plt.hist,"value",cumulative=False, bins=bins1)
g_lognor_melt.fig.suptitle("Log-normal\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_lognor_melt=sns.FacetGrid(df_lognor_melt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_lognor_melt.map(plt.hist,"value",cumulative=True, bins=bins1)
g_lognor_melt.fig.suptitle("Log-normal\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# The expit function, also known as the logistic sigmoid function, 
# is defined as expit(x) = 1/(1+exp(-x)). It is the inverse of the logit function.

df_expilt = expit(df_lognor).melt()
```

```python
# show the hist for the explit data
bins1 = 128

g_lognor_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_lognor_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_lognor_melt.fig.suptitle("Log-normal expit\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_lognor_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_lognor_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_lognor_melt.fig.suptitle("Log-normal expit\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# show the hist for the explit data
bins1 = 64

g_lognor_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_lognor_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_lognor_melt.fig.suptitle("Log-normal expit\nbins={}, cumulative False".format(bins1), y =1.15)

# show the hist for the explit data but with the cumulative mode ON
g_lognor_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=len(dict_with_data))
g_lognor_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_lognor_melt.fig.suptitle("Log-normal expit\nbins={}, cumulative True".format(bins1), y =1.15)
```

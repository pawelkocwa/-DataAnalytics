```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import logit, expit
```

```python
# Const to the part This is a code from ex2_1
SIZE = 100000
loc1 = 2
loc2 = 5
scale1=2
scale2=5

dict_with_data = {
    "r_nor":np.random.normal(size=SIZE),
    "r_nor_loc_{}_scale_{}".format(loc1, scale1):np.random.normal(size=SIZE, scale=scale1, loc=loc1),
    "r_nor_loc_{}_scale_{}".format(loc2, scale2):np.random.normal(size=SIZE, scale=scale2, loc=loc2)
}
# convert to the data frame object
df_nor=pd.DataFrame.from_dict(dict_with_data)
# melt data
df_nor_melt = df_nor.melt()
```

```python
# show the hist
bins1 = 128

g_nor_melt=sns.FacetGrid(df_nor_melt, col="variable", sharex=True, col_wrap=3)
g_nor_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_nor_melt.fig.suptitle("Without transforming\nbins={}, cumulative False".format(bins1), y =1.15)
```

```python
# The expit function, also known as the logistic sigmoid function, 
# is defined as expit(x) = 1/(1+exp(-x)). It is the inverse of the logit function.

df_expilt = expit(df_nor).melt()
```

```python
# show the hist for the explit data
bins1 = 128

g_nor_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=3)
g_nor_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_nor_melt.fig.suptitle("With transforming\nbins={}, cumulative False".format(bins1), y =1.15)
# show the hist for the explit data but with the cumulative mode ON
g_nor_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=3)
g_nor_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_nor_melt.fig.suptitle("With transforming\nbins={}, cumulative True".format(bins1), y =1.15)
```

```python
# show the hist for the explit data
bins1 = 64

g_nor_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=3)
g_nor_melt.map(plt.hist,"value",cumulative=False,bins=bins1)
g_nor_melt.fig.suptitle("With transforming\nbins={}, cumulative False".format(bins1), y =1.15)
# show the hist for the explit data but with the cumulative mode ON
g_nor_melt=sns.FacetGrid(df_expilt, col="variable", sharex=True, col_wrap=3)
g_nor_melt.map(plt.hist,"value",cumulative=True,bins=bins1)
g_nor_melt.fig.suptitle("With transforming\nbins={}, cumulative True".format(bins1), y =1.15)
```

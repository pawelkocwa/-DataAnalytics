```python
import pandas as pd
import matplotlib.pyplot as plt 
import math
import seaborn as sns
```

```python
# Import Data1.csv file to python.
data = pd.read_csv("Data1.csv")
data.head()
```

```python
# Set first column as the index.
data=data.set_index("Unnamed: 0")
```

```python
# Plot all columns as time series.
data.plot()
```

```python
# Plot histograms of all columns, verify bin size. Plot all on a single, faceted plot (use sns.FacetGrid and melt, use sharex=False for FacetGrid)
data.hist()
```

```python
# melt data
data_melt = data.melt()
data_melt.head()
```

```python
# histogram for melted data sns.FacetGrid
g = sns.FacetGrid(data_melt, col="variable")
g = g.map(plt.hist, "value")
```

```python
# Plot KDE-s (Kernel Denisty Estimators) for all columns.
sns.distplot(data_melt['value'],hist=False)
plt.show()
```

```python
# Repeat analysis for columns $\theta_1$-$\theta_4$ in 2018.

data_18  = data.loc['2018-01-01':'2018-12-31']
data_18.head()
```

```python
# melt data
data_melt_18 = data_18.melt()
data_melt_18.head()
```

```python
# histogram for melted data sns.FacetGrid
g = sns.FacetGrid(data_melt_18, col="variable")
g = g.map(plt.hist, "value")
```

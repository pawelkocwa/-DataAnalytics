```python
import pandas as pd
import matplotlib.pyplot as plt 
import math
import seaborn as sns
```

Comments for all exercise are in the end. The most needful are in the code.

```python
# Import Data1.csv file to python.
data = pd.read_csv("Data1.csv")
data.head()
```

```python
# some basic info about the data
data.info()
```

```python
# Set first column as the index. 
# Now Date is a index so we can refer to the data by it.
data=data.set_index("Unnamed: 0")
```

```python
# Plot all columns as time series.
data.plot()
```

```python
# Plot data in seperate subplots

# Due to the fact, plots looked bad, i have changed the size of them
data.plot(subplots=True, figsize=(10, 12))
```

```python
# Plot histograms of all columns, verify bin size. 
# Plot all on a single, faceted plot (use sns.FacetGrid and melt, 
# use sharex=False for FacetGrid)
data.hist()
```

```python
# melt data
data_melt = data.melt()
data_melt.head()
```

```python
# histogram for melted data with use of sns.FacetGrid
g = sns.FacetGrid(data_melt, col="variable")
g = g.map(plt.hist, "value")
```

```python
# Plot KDE-s (Kernel Denisty Estimators) for all columns.

# get names of all columns, but withoit first that is a date type
names = list(data.columns)[1:]

# automaticly plot KDE-s for all columns 
for i in names:
    plt.figure()
    sns.distplot(data[i], kde=True, hist=False)
```

```python
# Repeat analysis for columns $\theta_1$-$\theta_4$ in 2018.


# now I am doing everything the same way, but this time the data 
# range is a little bit smaller
data_18  = data.loc['2018-01-01':'2018-12-31']
data_18 = data_18[['theta_1', 'theta_2', 'theta_3', 'theta_4']]
data_18.head()
```

```python
# melt data
data_melt_18 = data_18.melt()
data_melt_18
```

```python
# histogram for melted data sns.FacetGrid
g = sns.FacetGrid(data_melt_18, col="variable", sharex=False, col_wrap=2)
g = g.map(plt.hist, "value")
```

```python
# Plot KDE-s (Kernel Denisty Estimators) for all columns.

# get names of all columns, but withoit first that is a date type
names = list(data_18.columns)[1:]

# automaticly plot KDE-s for all columns 
for i in names:
    plt.figure()
    sns.distplot(data_18[i], kde=True, hist=False)
```
